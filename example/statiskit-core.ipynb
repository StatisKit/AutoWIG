```python
from path import path
import os
rootdir = path(u'/home/pfernique/Desktop/StatisKit-Core.back')
for filepath in (rootdir/'src'/'wrapper').walkfiles('*.[hc]*'):
    os.remove(filepath.abspath())
headers = [str(header.abspath()) for header in (rootdir/'build-scons'/'include').walkfiles('*.h')]
```

```python
from autowig import autowig
import pickle
try:
    raise
    with open('statiskit-core.pkl', 'r') as f:
        asg = pickle.load(f)
except:
    asg = autowig.AbstractSemanticGraph()
    autowig.parser.plugin = 'pyclanglite'
    asg = autowig.parser(asg, headers, flags=['-ftemplate-depth-100', '-x', 'c++', '-std=c++11', '-stdlib=libc++',
                                              str('-I'+rootdir.abspath()/'build-scons'/'include')],
                         bootstrap=2)
    with open('statiskit-core.pkl', 'w') as f:
        pickle.dump(asg, f)
```

```python
autowig.controller.plugin = 'default'
asg = autowig.controller(asg)
```

```python
autowig.generator.plugin = 'boost_python_internal'
wrappers = autowig.generator(asg,
                             module=rootdir/'src'/'wrapper'/'_core.cpp',
                             decorator=rootdir/'src'/'statiskit'/'core'/'_core.py', closure=True)
for wrapper in wrappers:
    wrapper.write()
```

```python
session = autowig.scons(rootdir, '-k', '-j6')
session
```

```python
err = ''
feedback = autowig.feedback(session, rootdir, asg,
                            variantdir='build-scons',
                            tabsize=4, indent=0,
                            force=True)
while not session.err == err:
    err = session.err
    exec feedback
    session = autowig.scons(rootdir,
                            '-k', '-j1')
    feedback = autowig.feedback(session, rootdir, asg,
                            variantdir='build-scons',
                            tabsize=4, indent=0,
                            force=True)
session
```
