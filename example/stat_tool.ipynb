```python
from autowig import autowig
from path import path

rootdir = path('../..').abspath()
rootdir = rootdir/'StructureAnalysis'

includedir = rootdir + "/stat_tool/build-scons/include"
headers = [str(f) for f in includedir.walkfiles('*.h')]

flags = ['-x', 'c++', '-g', '-std=c++11', '-stdlib=libstdc++',
         '-DQT_GUI_LIB', '-DBOOST_PYTHON_DYNAMIC_LIB', '-DQT_SHARED', '-DMESSAGE',
         '-I/usr/include', '-I/usr/include/python2.7', 
         '-I/usr/local/openalea/include', '-I/usr/include/qt4',
         '-I/usr/include/qt4', '-I/usr/include/qt4/QtOpenGL',
         '-I/usr/include/qt4/QtGui', '-I/usr/include/qt4/QtCore'
         '-I' + str(rootdir + '/tool/build-scons/include'), '-I' + str(includedir)]
```

```python
asg = autowig.AbstractSemanticGraph()
autowig.parser.plugin = 'pyclanglite'
autowig.parser(asg, headers, flags=flags)
import pickle
with open('stat_tool.pkl', 'w') as f:
    pickle.dump(asg, f)
```

```python
import pickle
with open('stat_tool.pkl', 'r') as f:
    asg = pickle.load(f)
from autowig import autowig
```

```python
 asg['class ::stat_tool::StatInterface'].declarations()
```

```python
interface = asg['class ::stat_tool::StatInterface']
interfaces = [cls.globalname for cls in interface.subclasses()]
for cls in asg.classes(pattern='.*::stat_tool::.*'):
    if not cls.globalname in interfaces:
        cls.boost_python_export = False
for cls in asg.classes(pattern='.*::stat_tool::.*Plot.*'):
    cls.boost_python_export = True
interface.boost_python_export = True
asg['class ::stat_tool::StatError'].boost_python_export = True
```

```python
autowig.generator.plugin = 'boost_python_internal'
autowig.generator(asg, module=rootdir/'stat_tool'/'src'/'wrapper'/'__stat_tool.cpp', decorator=None)
```

```python
for export in asg[rootdir/'stat_tool'/'src'/'wrapper'/'__stat_tool.cpp'].exports:
    export.write()
```
